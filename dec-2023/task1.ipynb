```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import librosa
import librosa.display
import soundfile as sf
import os
import IPython.display as ipd
```

```python
x, sr = librosa.load(os.path.join('files', 'Task_1', 'T1_audio.wav'))
x.shape, sr
```

```python
x, sr = sf.read(os.path.join('files', 'Task_1', 'T1_audio.wav'))
x.shape, sr
```

```python
# https://github.com/librosa/librosa/issues/1763
plt.figure(figsize=(14, 5))
librosa.display.waveshow(x, sr=sr, color='blue')
```

```python
X = librosa.stft(x)
Xdb = librosa.amplitude_to_db(abs(X))
plt.figure(figsize=(14, 5))
librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='hz')
```

```python
ipd.Audio(os.path.join('files', 'Task_1', 'T1_audio.wav'))
```

```python
# https://pysoundfile.readthedocs.io/en/0.8.1/#soundfile.write
sf.write('test.wav', x, sr)
```

```python
x_fast = librosa.effects.time_stretch(x, rate=2.0)
sf.write('task1.wav', x_fast[::-1], sr)
ipd.Audio('task1.wav')
```
